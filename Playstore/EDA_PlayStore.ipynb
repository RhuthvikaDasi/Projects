## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

#Removing dulplicates from dataset
app_data.drop_duplicates(inplace=True)

# Converting 'Rating' and 'Reviews' to float
app_data['Rating'] = app_data['Rating'].astype(float)
app_data['Reviews'] = app_data['Reviews'].astype(float)

# Converting 'size' to float
app_data['Size'] = app_data['Size'].astype(str)  # Converting size to string
app_data['Size'] = app_data['Size'].str.replace('M', '000')  # Converting M to Kbytes
app_data['Size'] = app_data['Size'].replace('Varies with device', np.nan)
app_data['Size'] = app_data['Size'].str.replace('k', '')
app_data['Size'] = app_data['Size'].replace('1,000+', '1000')
app_data['Size'] = app_data['Size'].astype(float)

# Cleaning 'Price' column
app_data['Price'] = app_data['Price'].astype(str)
app_data['Price'] = app_data['Price'].str.replace('$', '')
app_data['Price'] = app_data['Price'].replace('Everyone', np.nan)
app_data['Price'] = app_data['Price'].astype(float)

# Cleaning 'Installs' column
app_data['Installs'] = app_data['Installs'].astype(str)
app_data['Installs'] = app_data['Installs'].str.replace(',', '')
app_data['Installs'] = app_data['Installs'].str.replace('+', '')
app_data['Installs'] = app_data['Installs'].replace('Free', np.nan)
app_data['Installs'] = app_data['Installs'].astype(float)

# Converting Sentiment_Polarity and Sentiment_Subjectivity to float
app_data['Sentiment_Polarity'] = app_data['Sentiment_Polarity'].astype(float)
app_data['Sentiment_Subjectivity'] = app_data['Sentiment_Subjectivity'].astype(float)

app_data['Rating'].fillna(app_data['Rating'].median(), inplace=True)
app_data['Size'].fillna(app_data['Size'].median(), inplace=True)
app_data['Translated_Review'].fillna('', inplace=True)
app_data['Sentiment'].fillna('Neutral', inplace=True)
app_data['Sentiment_Polarity'].fillna(0, inplace=True)
app_data['Sentiment_Subjectivity'].fillna(0, inplace=True)


# Check for any remaining missing values
print(app_data.isnull().sum())
````

### What all manipulations have you done and insights you found?

There are missing values and dupicates present in the dataset.The manipulations I have done is first removed duplicates from dataset.Converted ratings,reviews,sentiment_polarity,sentiment_subjectivity,size and also cleaned the prices and installs data
